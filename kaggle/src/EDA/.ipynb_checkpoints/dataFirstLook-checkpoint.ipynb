# First Look at the Dataset
This file is to collect the training images and copy them into one single file. So we can actually see the data and have a better understanding of the pictures. Then we have a first look at the data set. We them display different types of the images with the following focues:
- color
- numbers of nuclei
- size and shape
- possible noise

## File Reconstruct

```python
import numpy as np
import pandas as pd
import os
import glob
import math
from os.path import join
import shutil

ROOT_PATH = '../../'
TRAIN_PATH = ROOT_PATH + 'data/stage1_train/'
# TEST_PATH = ROOT_PATH + 'data/stage1_test/'

# read file names
train_ids = os.listdir(TRAIN_PATH)
# test_ids = os.listdir(TEST_PATH)

# read image paths
train_image_paths = [glob.glob(join(TRAIN_PATH, train_id, "images", "*"))[0] for train_id in train_ids]
```

# Copy the images into destination file

```python
N = len(train_image_paths)

desDir = ROOT_PATH + 'result/train_imgs/'

for i in range(N):
    desPath = desDir + str(i+1) + '.png'
    shutil.copyfile(train_image_paths[i], desPath)
    
print ("Please go to 'root/result/train_imgs' directory to see the training images.")
print ("The images are stored in numbers (starting from 1), and are stored in png format.")
```

# First glance of the images
First we look at these images to have a first impression of the images. Here are some conclusions of the images.

## 1. colors
Based on color, the images can be sorted into three classes.
- Colored images with whtie background.
- Colored images with purple background.
- B&W images with black background.
- B&W images with white background.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
plt.figure(figsize = (9,9))

filePath = desDir


file = filePath + '3.png'
img=mpimg.imread(file)
plt.subplot(221)
plt.title('color image (white background)')
imgplot = plt.imshow(img)

file = filePath + '53.png'
img=mpimg.imread(file)
plt.subplot(222)
plt.title('color image - (black background)')
imgplot = plt.imshow(img)

file = filePath + '42.png'
img=mpimg.imread(file)
plt.subplot(223)
plt.title('B&W (black background)')
imgplot = plt.imshow(img)

file = filePath + '113.png'
img=mpimg.imread(file)
plt.subplot(224)
plt.title('B&W (white background)')
imgplot = plt.imshow(img)

plt.savefig('differentColor_imgs.png');
```

## 2. numbers
Number of the nuclei on every image varies. Some pictures only have few nucleis, while some have hundreds of nucleis.

```python
file = filePath + '603.png'
img=mpimg.imread(file)
plt.subplot(121)
plt.title('Very few nucleis')
imgplot = plt.imshow(img)

file = filePath + '521.png'
img=mpimg.imread(file)
plt.subplot(122)
plt.title('Large amount of nucleis')
imgplot = plt.imshow(img)

plt.savefig('differentNumber_imgs.png');
```

According to the images, we identify four different types of the noises:

The image below shows the four different tyoes of the noises.

```python
plt.figure(figsize = (8,8))

file = filePath + '284.png'
img=mpimg.imread(file)
plt.subplot(221)
plt.title('Blurring Image')
imgplot = plt.imshow(img)

file = filePath + '583.png'
img=mpimg.imread(file)
plt.subplot(222)
plt.title('Low Constract Image')
imgplot = plt.imshow(img)

file = filePath + '463.png'
img=mpimg.imread(file)
plt.subplot(223)
plt.title('Low Resolution Image')
imgplot = plt.imshow(img)

file = filePath + '238.png'
img=mpimg.imread(file)
plt.subplot(224)
plt.title('Background Noise')
imgplot = plt.imshow(img)

plt.suptitle('Nuclei with 4 types of possible noises')

plt.savefig('differentNumber_Noises.png');
```
